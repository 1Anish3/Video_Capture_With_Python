```python
import cv2,time
```

```python
video = cv2.VideoCapture(0)
```

```python
a=0
```

```python
while True:
    
    #Calculating number of iterations
    a=a+1
    
    #Creating a boolean and an ndArray
    check,frame = video.read()
    print(check)
    print(frame)
    
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    #time.sleep(3)
    
    cv2.imshow("Capture",gray)
    
    key=cv2.waitKey(1)
    
    if key==ord('q'):
        break
        
print(a)      
video.release()
cv2.destroyAllWindows()
```

```python
#Total number of iterations in 3 seconds is 74
#i.e., Approximately 24 / 25 fps(frames per second)
```
